```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
VARD = 50
```

```python
image = cv2.cvtColor(cv2.imread('img2.png'), cv2.COLOR_BGR2RGB )
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
r, g, b = cv2.split(image)

r, g, b = r / 255, g / 255, b / 255
```

```python
pcaComponents = VARD

pca_R = PCA(n_components = pcaComponents)
reduced_R = pca_R.fit_transform(r)

pca_G = PCA(n_components = pcaComponents)
reduced_G = pca_G.fit_transform(g)

pca_B = PCA(n_components = pcaComponents)
reduced_B = pca_B.fit_transform(b)
```

```python
combined = np.array([reduced_R, reduced_G, reduced_B])
```

```python
reconstructed_R = pca_R.inverse_transform(reduced_R)
reconstructed_G = pca_G.inverse_transform(reduced_G)
reconstructed_B = pca_B.inverse_transform(reduced_B)
```

```python
image_reconstructed = (cv2.merge((reconstructed_R, reconstructed_G, reconstructed_B)))
```

```python
plt.imshow(image_reconstructed)
```
